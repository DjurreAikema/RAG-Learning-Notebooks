# 🧪 RAG Bot Test 1: Conversational Memory

This is a **playground** notebook. It assumes the
`RAG Bot basics tutorial.ipynb` is done and the vector DB is already built and synced (via
`Sync Vector DB.ipynb`). Nothing gets loaded, split, or embedded here. It just **connects to the
existing `chroma_db/`** and experiments with one new capability: **chat memory**, for a
real back-and-forth with the bot about the documents.

### What's new here (and what isn't)

Already covered elsewhere, so no re-explaining: loading docs, chunking, embeddings, building the
Chroma DB. The Sync notebook owns all of that now.

**New in this notebook**:
- **Conversational memory**: the bot remembers what was said earlier in the session.
- **History-aware retrieval**: the hard part of memory in RAG. A follow-up like *"what about for
  members?"* is meaningless to the retriever on its own. It gets rewritten into a standalone question
  *using the chat history* **before** hitting the vector DB, so retrieval still finds the right chunks.
- **The modern LangChain "retrieval chain" helpers** (`create_history_aware_retriever`,
  `create_stuff_documents_chain`, `create_retrieval_chain`): these replace the hand-wired LCEL
  chain from the tutorial and are built for exactly this memory use case.

### The problem, in one picture

```
   You: "What's the refund policy?"        → retriever gets a clear question → good chunks ✅
   You: "What about for members?"          → retriever sees only "what about for members?"
                                              → no idea what "what about" refers to → junk chunks ❌

   FIX: rewrite using history FIRST:
   ("What about for members?" + history) → LLM → "What is the refund policy for members?"
                                              → retriever gets a clear question again → good chunks ✅
```

## Step 1: Configuration

Same constants as the tutorial and the Sync notebook. They **must** match, since this reads the
exact collection those notebooks wrote. Nothing is embedded here, but `EMBED_MODEL` still has to be
identical because Chroma re-embeds the live questions with it at query time. A mismatched embedder
would search the wrong vector space.

````python
from pathlib import Path

# --- Models (must match the tutorial / Sync notebook) ---
EMBED_MODEL = "nomic-embed-text"   # embeds questions at query time
LLM_MODEL   = "llama3.1:8b"        # writes the answers (and rewrites follow-ups)

# --- Where the already-built DB lives ---
DB_DIR          = "../chroma_db"
COLLECTION_NAME = "company_docs"

# --- Retrieval knob: how many chunks to pull per question ---
N_RESULTS = 4

print("Config loaded.")
print("  DB dir    :", Path(DB_DIR).resolve())
print("  Collection:", COLLECTION_NAME)
print("  LLM       :", LLM_MODEL)
print("  Embedder  :", EMBED_MODEL)
````

## Step 2: Connect to the existing vector DB

This is the whole point of leaning on the Sync notebook: **no indexing here.** Just reopen the
persisted Chroma collection and wrap it in a retriever. If this cell reports 0 vectors, the DB
isn't built yet. Go run `Sync Vector DB.ipynb` first.

````python
from langchain_ollama import OllamaEmbeddings
from langchain_chroma import Chroma

embeddings = OllamaEmbeddings(model=EMBED_MODEL)

vectorstore = Chroma(
    persist_directory=DB_DIR,
    embedding_function=embeddings,
    collection_name=COLLECTION_NAME,
)

count = vectorstore._collection.count()
print(f"Reconnected to '{COLLECTION_NAME}': {count} vectors.")
assert count > 0, "DB is empty. Run the Sync Vector DB notebook first."

retriever = vectorstore.as_retriever(search_kwargs={"k": N_RESULTS})
print("Retriever ready.")
````

## Step 3: The LLM

One `ChatOllama` instance does double duty in this notebook: it **rewrites** follow-up questions and
it **writes** the final answers. `temperature=0` keeps both jobs focused and repeatable.

````python
from langchain_ollama import ChatOllama

llm = ChatOllama(model=LLM_MODEL, temperature=0)
print("LLM ready.")
````

## Step 4: History-aware retrieval (the core new idea)

Here's the piece that makes memory actually *work* in RAG.

A normal retriever takes a question string and finds similar chunks. But in a conversation, the
user's latest turn is often not self-contained, *"what about for members?"*, *"and the deadline?"*,
*"why?"*. Embedding those directly retrieves garbage, because the meaning lives in the **earlier
turns**, not in the words themselves.

`create_history_aware_retriever` solves this by inserting an LLM step *in front of* the retriever:

1. It shows the LLM the **chat history + the new question**.
2. It asks the LLM to **reformulate** the new question into a standalone one that makes sense on its
   own (this is what the `contextualize` prompt below instructs).
3. That rewritten question (not the raw user text) is what gets embedded and sent to Chroma.

One subtlety worth calling out: the prompt tells the model to return the question **unchanged** if it
already stands on its own. So the very first question, or any fully-specified question, passes
straight through without a pointless rewrite.

`MessagesPlaceholder("chat_history")` is new too, it's a slot in the prompt where we'll drop the
running list of past messages (as `HumanMessage`/`AIMessage` objects) at call time.

````python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableBranch

contextualize_system_prompt = (
    "Given the chat history and the latest user question, which might reference context "
    "in the chat history, formulate a standalone question that can be understood without "
    "the chat history. Do NOT answer the question, just reformulate it if needed, and "
    "otherwise return it unchanged."
)

contextualize_prompt = ChatPromptTemplate.from_messages([
    ("system", contextualize_system_prompt),
    MessagesPlaceholder("chat_history"),
    ("human", "{input}"),
])

contextualize_chain = contextualize_prompt | llm | StrOutputParser()

history_aware_retriever = RunnableBranch(
    (
        lambda x: not x.get("chat_history"),
        (lambda x: x["input"]) | retriever,
    ),
    contextualize_chain | retriever,
)

print("History-aware retriever ready (LCEL, no langchain-classic).")
````

## Step 5: The answering prompt

This is the tutorial's grounded, "answer only from context, admit when you don't know" prompt with
two changes for conversation:

- A `MessagesPlaceholder("chat_history")` so the model can see the dialogue so far (useful for tone
  and for resolving references the retrieval step didn't fully capture).
- `{context}` is still where retrieved chunks land, and `{input}` is the current question. We use
  `{input}` rather than the tutorial's `{question}` because the retrieval-chain helpers below expect
  that key name.

````python
qa_system_prompt = (
    "You are a helpful assistant for answering questions about internal company documents. "
    "Use ONLY the pieces of context provided below to answer the question. "
    "If the answer is not contained in the context, say you don't know based on the available "
    "documents do not invent information. Keep answers concise and cite which document you used."
    "\n\nContext:\n{context}"
)

qa_prompt = ChatPromptTemplate.from_messages([
    ("system", qa_system_prompt),
    MessagesPlaceholder("chat_history"),
    ("human", "{input}"),
])

print("QA prompt ready.")
````

## Step 6: Assemble the conversational chain

The tutorial hand-wired retrieval → prompt → LLM with the `|` pipe. For memory we use two purpose-built
helpers instead, then glue them together:

- **`create_stuff_documents_chain(llm, qa_prompt)`**: takes retrieved documents, "stuffs" them into
  the `{context}` slot of `qa_prompt`, and runs the LLM. (It also handles formatting the docs, so we
  no longer need the tutorial's manual `format_docs`.)
- **`create_retrieval_chain(history_aware_retriever, question_answer_chain)`**: wires the
  history-aware retriever to the answering chain. Call it with `{"input": ..., "chat_history": ...}`
  and it returns a dict with the `answer` **and** the retrieved `context`, so we can still show sources.

The result is one object that does: *rewrite question → retrieve → answer*, all memory-aware.

````python
from langchain_core.runnables import RunnablePassthrough

def format_docs(docs):
    # Combine retrieved chunks into one context string, tagged by source (like the tutorial).
    blocks = []
    for d in docs:
        src = d.metadata.get("source", "unknown")
        blocks.append(f"[Source: {src}]\n{d.page_content}")
    return "\n\n---\n\n".join(blocks)

# Full conversational chain, assembled by hand:
#   1) run the history-aware retriever, stash the retrieved docs under "context"
#   2) format those docs into a string JUST for the prompt, then answer
# The outer "context" stays a LIST of docs, so chat() can still list sources;
# only the inner copy handed to the prompt is flattened to text.
rag_chain = (
    RunnablePassthrough.assign(context=history_aware_retriever)
    | RunnablePassthrough.assign(
        answer=(
            RunnablePassthrough.assign(context=lambda x: format_docs(x["context"]))
            | qa_prompt
            | llm
            | StrOutputParser()
        )
    )
)

print("Conversational RAG chain assembled (LCEL).")
````

## Step 7: Where the memory actually lives

The chain itself is **stateless**. It doesn't remember anything between calls. The "memory" is just
a plain Python list of messages that *we* keep and pass in every time, appending each new exchange.

For a single notebook session, a bare list is the simplest thing that works (no need for LangChain's
heavier `RunnableWithMessageHistory` / session-store machinery). The user's turn gets added as a
`HumanMessage` and the bot's reply as an `AIMessage` after each call.

The helper below does one round-trip: send `chat_history + question`, print the answer and its
sources, then record both turns into `chat_history`.

````python
from langchain_core.messages import HumanMessage, AIMessage

chat_history = []   # our running memory for this session

def chat(question: str, show_sources: bool = True):
    result = rag_chain.invoke({"input": question, "chat_history": chat_history})
    answer = result["answer"]

    print("🧑 You:", question)
    print("🤖 Bot:", answer)

    if show_sources:
        seen = set()
        srcs = [d.metadata.get("source", "unknown") for d in result["context"]]
        uniq = [s for s in srcs if not (s in seen or seen.add(s))]
        print("📄 Sources:", ", ".join(uniq) if uniq else "(none)")

    # Record BOTH turns so the next question has context to work with.
    chat_history.extend([
        HumanMessage(content=question),
        AIMessage(content=answer),
    ])
    return answer

def reset_memory():
    chat_history.clear()
    print("🧹 Memory cleared.")

print("Helpers ready: chat(...) and reset_memory().")
````

## Step 8: Try a follow-up by hand first

Before the loop, run these two cells in order to *see* memory working. The second question only makes
sense because of the first. If history-aware retrieval is doing its job, the bot will understand it.

Swap in questions that fit the documents.

````python
reset_memory()
_ = chat("What is the company's refund policy?")
````

````python
# Notice: this question is NOT self-contained. Memory + rewriting make it work.
_ = chat("And what about for members?")
````

## Step 9: Interactive chat loop

Now the real back-and-forth. This keeps looping, threading `chat_history` through every turn, so
genuine follow-ups work. Type `quit` / `exit` to stop, or `reset` to wipe the memory and start a
fresh conversation without restarting the kernel.

> Tip: watch how it handles a chain of references. For example *"What's the vacation policy?"* →
> *"How much carries over?"* → *"And for part-timers?"*. Each depends on the last.

````python
print("Chat with your documents. Commands: 'quit' / 'exit' to stop, 'reset' to clear memory.\n")
while True:
    try:
        q = input("You: ").strip()
    except (EOFError, KeyboardInterrupt):
        break
    if q.lower() in {"quit", "exit"}:
        print("Goodbye!")
        break
    if q.lower() == "reset":
        reset_memory()
        continue
    if not q:
        continue
    chat(q)
    print()
````

## 🎉 What this test demonstrated

- **Loaded, didn't build**: the DB came straight from the Sync notebook. This notebook never
  indexes anything.
- **Memory = a list we pass in**: the chain stays stateless. We own the `chat_history`.
- **History-aware retrieval**: follow-ups get rewritten into standalone questions *before* hitting
  Chroma, which is what makes conversational RAG actually retrieve the right chunks.
- **Modern retrieval-chain helpers**: `create_history_aware_retriever` +
  `create_stuff_documents_chain` + `create_retrieval_chain` replaced the hand-wired LCEL pipeline.
